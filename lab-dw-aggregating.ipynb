```python
#2. Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for 
#customers who have responded "Yes" to the last marketing campaign. Write your conclusions.
data_2=data[data['Response']=='Yes']
pd.pivot_table(data_2, index=['Policy Type'], columns=['Gender'], values=['Total Claim Amount'])
#females have a higher.mean claim amount versus the males. But only the personal policy type Auto the males have an little higher main than the females
```

```python
#4. Find the maximum, minimum, and median customer lifetime value by education level and gender. 
#Write your conclusions
data.pivot_table(index='Education',columns='Gender',values=['Customer Lifetime Value'], aggfunc=['max','min','mean','std'])

#there must be significant values that make the mean, being a bit apart.
```

```python
#5. The marketing team wants to analyze the number of policies sold by state and month. Present the data in a table
#where the months are arranged as columns and the states are arranged as rows.
data_df=data
data_df.State.fillna("no state mentioned", inplace=True)
```

```python
data_5=data_df.pivot_table(index='State',columns='Months Since Policy Inception',values=['Number of Policies'], aggfunc=['sum'])
```

```python
data_5['total']=data_5.sum(axis=1)
```

```python
data_5df=pd.DataFrame(data_5)
```

```python
data_5df
```

```python
data_5df['total'].sum()==data_df['Number of Policies'].sum() # to confirm that all are included in the DataFrame done
```

```python
#6. Display a new DataFrame that contains the number of policies sold by month, by state, for the top 3 states with 
#the highest number of policies sold.
new_dataF6=data_5df.sort_values(by=['total'],ascending=False)
new_dataF6
```
